# df.astype()
```python
df.astype(int,float,str,...)
```

df.info()에서 보여주는 데이터 타입 별로 설명:
- object:
텍스트 또는 혼합된 데이터 타입.
주로 문자열 데이터에 사용됨.
- int64:
정수형 데이터.
- float64:
부동소수점 숫자.
- bool:
부울 값 (True 또는 False).
참/거짓 값을 나타냄.
- datetime64:
날짜 및 시간 데이터.
날짜와 시간 정보 처리에 사용됨.
- category:
범주형 데이터.
제한된 수의 고유값들을 가지며 순서가 없는 그룹 또는 분류를 나타냄.

````python
import pandas as pd
data = {'A': ['1', '2', '3'],
        'B': ['4.5', '5.6', '6.7'],
        'C': ['True', 'False', 'True']}
df = pd.DataFrame(data)
df
````

````python
# 덮어쓰기 해야함
df["A"] = df["A"].astype(int)
df["B"] = df["B"].astype(float)
df["C"] = df["C"].astype(bool)
df.info()
````

#na
- na: 누락된 데이터를 나타내는 일반적인 용어
  - NaN: 부동 소수점 데이터셋에서 누락된 값
  - None: 객체 타입(예: 문자열) 데이터셋에서 '값이 없음'을 나타내는 값

````python
import pandas as pd
import numpy as np

# 숫자계열에 있으면 None는 NaN 범주형계열은 None로 출력됨
# ''은 None과 다름
data = {'숫자열': [1, 2, None, 4,None],
        '문자열': ['사과', None, '바나나', '체리','']}

df = pd.DataFrame(data)
df
````

````python
# ''은 빈값으로 취급하지 않는다.
df.isna()
````

# dropna()
```python
dropna(subset=['col_1','col_2',...], how = 'any/all')
```
- 누락된 데이터를 포함하는 행이 제거
- how 매개변수:

  - how='any': 하나라도 None이 있는 행 제거 (기본값)
  - how='all': 모든 값이 None인 행만 제거

````python
import pandas as pd
import numpy as np

# 샘플 데이터 프레임 생성
data = {'A': [1, None, 3, 4, None, None, 10],
        'B': [None, 2, 3, None, 5, None, 10],
        'C': [1, 2, None, 4, 5, None, 10 ]}

df = pd.DataFrame(data)
df
````

````python
# NaN값을 포함하지 않은 데이터만 출력
df.dropna()
````

````python
# 모든 값이 NaN인 데이터는 빼고 출력
df.dropna(how = 'all')
````

````python
# A에 NaN인 경우 빼고 출력
df.dropna(subset=["A"], how='any')
````

# fillna()
```python
df.fillna(
  {'col_1' : value1
  , 'col_2' : value2}
  , inplace = True
)
```

````python
# drop na df
import pandas as pd
import numpy as np

# 샘플 데이터 프레임 생성
data = {'A': [1, None, 3, 4, None, None, 10],
        'B': [None, 2, 3, None, 5, None, 10],
        'C': [1, 2, None, 4, 5, None, 10 ]}

df = pd.DataFrame(data)
##

df
````

````python
df.fillna(0) #inplace=True 해야 덮어써짐
````

````python
df.fillna(
  {'A' : 1
  , 'B' : 100
  , 'C' : 1000}
  , inplace = True
)
df
````

````python
df.fillna(
  {'A' : df.A.mean()
  , 'B' : df.B.sum()
  , 'C' : df.C.max()}
  )
````

# to_numeric()
- 용도: 문자열, 다른 타입 데이터 → 숫자형 데이터 변환
- 에러 처리 옵션: errors 인자 사용
- 사용 예: '123', '3.14' 같은 문자열 → 정수 123, 실수 3.14 변환
- errors='coerce': 변환 불가능한 값을 NaN으로 설정



````python
import pandas as pd

# numeric valeu 와 non-numeric value가 혼재된 상황
data = {'Value': ['10', '20', '30', 'error', '50',10]}
df = pd.DataFrame(data)

# Value 칼럼을 Numeric으로 변환, non-numeric value를 NaN 으로 처리
df.info() , df
````
