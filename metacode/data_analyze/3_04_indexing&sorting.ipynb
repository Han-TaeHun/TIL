# Index
* 인덱스(Index): 행을 구별하는데 사용되는 고유한 식별자.

````python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("taxis")
````

````python
df.head()
````

````python
# DataFrame에서 하나 이상의 열을 새로운 인덱스로 설정하는 데 사용
# color coplumn이 위에 DataFrame과는 다르게 인덱스칸으로 들어감
df.set_index('color', inplace=True)
````

````python
df.head()
````

# Sort
- 특정 기준에 따라 순서대로 `나열`
- sort_value():특정 column 기준
```python
df.sort_values('col1', ascending = False, inplace = True )
```
  - ascending = False: 내림차순
  - ascending = True: 오름차순

- sort_index(): index 기준
```python
df.sort_index(inplace= True)
```

````python
df.sort_values("distance", ascending=False, inplace=True)
````

````python
df.head(10)
````

````python
# 숫자 인덱스로 원복
df.reset_index(inplace=True)
````

````python
import pandas as pd

# 데이터프레임 생성
data = {
    'ProductID': [101, 102, 103, 104],
    'SaleDate': ['2021-07-01', '2021-07-03', '2021-07-02', '2021-07-04'],
    'Quantity': [3, 2, 5, 4]
}

df = pd.DataFrame(data)
````

````python
df.sort_values("ProductID", inplace=True)
df
````

````python
df.sort_values("Quantity", ascending=False, inplace=True)
df
````

````python
df.sort_index(inplace=True)
df
````
