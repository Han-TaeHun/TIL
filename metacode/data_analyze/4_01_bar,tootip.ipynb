````python
# 예시 데이터 생성
import pandas as pd

data = {
    'Month': ['January', 'February', 'March', 'January', 'February', 'March', 'January', 'February', 'March'],
    'Sales': [20000, 24000, 30000, 15000, 16000, 22000, 180000, 26000, 27000],
    'Ratio': [.234, .324, .3, .232, .1216, .332, .330, .330, .27],
    'Region': ['North', 'North', 'North', 'South', 'South', 'South', 'East', 'East', 'East']
}



df1 = pd.DataFrame(data)
df1
````

````python
import plotly.express as px

# 막대 그래프 생성
fig = px.bar(df1
    , x="Sales"  # x축 데이터
    , y="Month"  # y축 데이터
    , color="Region"  # 막대 색상 구분, 'col_3' 값에 따라 달라짐
    , title="Sale by Month"  # 그래프 제목 설정
    # , barmode='group'  # 막대 표시 방식, 'group'은 나란히, 'stack'은 쌓아서 표시
)
fig.update_layout(xaxis_title="매출", yaxis_title="월")
fig.show()
````

````python
df1_agg = df1.groupby(by="Month")["Sales"].sum().reset_index()
# 막대 그래프 생성
fig = px.bar(df1_agg
    , x="Sales"  # x축 데이터
    , y="Month"  # y축 데이터
    # , color="col_3"  # 막대 색상 구분, 'col_3' 값에 따라 달라짐
    , title="Sale by Month"  # 그래프 제목 설정
    , barmode='group'  # 막대 표시 방식, 'group'은 나란히, 'stack'은 쌓아서 표시
)

fig.show()
````

````python
import pandas as pd
import numpy as np
import plotly.express as px

# 샘플 데이터 생성
np.random.seed(0)
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
regions = ['North', 'South', 'East', 'West']

data = {# 이 코드는 이해하지 못하셔도 됩니다.
    'Month': np.random.choice(months, 100, replace=True),
    'Sales': np.random.randint(200, 1000, 100),
    'Profit': np.random.randint(20, 100, 100),
    'Region': np.random.choice(regions, 100, replace=True)
}

df = pd.DataFrame(data)
````

````python
df.head()
````

````python

# 막대 그래프 생성
fig = px.bar(df
    , x="Month"  # x축 데이터
    , y="Sales"  # y축 데이터
    , color="Region"  # 막대 색상 구분, 'col_3' 값에 따라 달라짐
    , title="월별 및 지역별 매출 합계"  # 그래프 제목 설정
    , barmode='group'  # 막대 표시 방식, 'group'은 나란히, 'stack'은 쌓아서 표시
)

# X축과 Y축 제목 설정 (선택적)
# fig.update_layout(xaxis_title="x_title_name", yaxis_title="y_title_name")

# 그래프 화면에 표시
fig.show()
````

````python
df.groupby(by=["Month", "Region"]).size()
````

````python
df_agg = df.groupby(by=["Month", "Region"])["Sales"].sum().reset_index()
df_agg["Profit_Margin"] = df_agg["Sales"] / df["Profit"]


# 막대 그래프 생성
fig = px.bar(df_agg
    , x="Month"
    , y="Profit_Margin"
    , color="Region"
    , title="월별 및 지역별 매출 합계"
    , barmode='group'
    , hover_data ={
    "Region" : False,
    "Profit_Margin" : ":.2f",}
)

# X축과 Y축 제목 설정 (선택적)
# fig.update_layout(xaxis_title="x_title_name", yaxis_title="y_title_name")

# 그래프 화면에 표시
fig.show()
````

````python
df_agg2 = df.groupby(["Month", "Region"]).agg(
    Sales_sum = ( "Sales", "sum"),
    Profit_sum = ( "Profit", "sum")
).reset_index()

df_agg2["profit_ratio"] = df_agg2.Profit_sum / df_agg2.Sales_sum
df_agg2
````

````python
# 막대 그래프 생성
fig = px.bar(df_agg2
    , x="Month"
    , y="Sales_sum"
    , color="Region"
    , title="월별 및 지역별 매출 합계"
    , barmode='group'
    , hover_data ={
    "Region" : False,
    "profit_ratio" : ":.2f",}
)

# X축과 Y축 제목 설정 (선택적)
# fig.update_layout(xaxis_title="x_title_name", yaxis_title="y_title_name")

# 그래프 화면에 표시
fig.show()
````
