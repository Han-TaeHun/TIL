# Data Type

* Categorical Data (범주형 데이터): 제한되 수의 범주로 나눌 수 있는 데이터. 예시로는 성별, 혈액형, 학년이 있음

* Numeric Data (수치데이터): 수치로 표현되며 계산이 가능한 데이터.
    * Binary Data(이진 데이터): '0'과 '1'로만 구성된 데이터. 예시로 합격여부,  물품 구매 여부 등이 있음, 이는 경우에 따라 범주형 데이터로도 분류.
    * Ratio Data (비율 데이터): 값들 사이의 비율 비교가 가능한 데이터. 예시로는 키, 몸무게, 가격
