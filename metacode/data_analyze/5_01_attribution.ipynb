# Attribution
- Attribution은 고객의 구매나 전환에 기여한 각 마케팅 접점의 영향력을 평가하는 방법입니다.

| Attribution Model       | 정의                                                  | 장점                                                 | 단점                                       |
|-------------------------|------------------------------------------------------|------------------------------------------------------|--------------------------------------------|
| **Last Click Attribution** | 마지막 클릭, 즉 구매 직전 접점에 모든 공을 돌림        | - 이해 및 실행이 쉬움<br>- 데이터 추적 용이           | - 초기 및 중간 단계의 중요한 접점 무시 가능 |
| **First Click Attribution** | 첫 클릭, 즉 고객 여정의 첫 접점에 모든 공을 돌림        | - 브랜드 인지도 및 초기 고객 유입 촉진 캠페인 평가에 적합 | - 구매 결정에 영향을 미친 후속 접점 간과 가능 |
| **Multi-Touch Attribution** | 전체 고객 여정에 걸쳐 여러 접점에 공을 분배            | - 고객의 전체 구매 결정 과정을 폭넓게 고려<br>- 다양한 접점의 복합적 영향 이해에 도움 | - 구현 복잡<br>- 적절한 가중치 결정 어려움 |

```python
#  visit과 purchase 의 cross join을 하여 가능한 모든 경우를 만든다.
# 그중에서 시간이 가장 가까운 것을 찾아 라스트 클릭을 찾아낸다.
df_purc = df[df.event == "purchase"]
df_purc
df_visit = df[~df.campaign.isna()]
df_visit
df_cross = pd.merge(df_purc, df_visit, how="cross")
df_cross.head()
```

```python
# id가 같은 것만 필요하다.
df_cross = df_cross[df_cross.user_id_x == df_cross.user_id_y]
```

```python
# 방문 시간보다 늦게 구매한 데이터가 필요함
df_cross = df_cross[df_cross.timestamp_x >= df_cross.timestamp_y]
df_cross[df_cross.user_id_x == 8]
```

```python
#  과거 방문 기록중에서 가장 가까운 기록이 필요함
df_cross['td'] = (df_cross.timestamp_x - df_cross.timestamp_y).dt.total_seconds()
```

```python
idx = df_cross.groupby(by='event_id_x').td.idxmin()
```
