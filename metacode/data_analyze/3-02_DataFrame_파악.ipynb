Dictonary

```python
my_dict = {
    'col1': [1, 2, 3, 4] ,
    'col2': [10, 20, 30, 40] ,
    'col3': [100, 200, 300, 400]
}
my_dict
```

```python
import pandas as pd

# 딕셔너리의 key 값은 columns가 되고 value 값은  value가 된다. 단, 리스트 안에 들어 있는 경우
# 단일 값일 때는 scala로 처리 되어 에러가 났음 Series로 처리 해야될 걸로 예상됨
df = pd.DataFrame(my_dict)
df
```
