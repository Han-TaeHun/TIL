## columns 변경


```python
# inplace =True로 설정하면 실제 데이터에 바로 적용
df_iris.rename( columns = {"sepal_width": "A", "petal_length": "B"}, inplace =True)
df_iris
```

##  column 추가

```python
df_iris["A_plus_B"] =  df_iris.A + df_iris.B
df_iris
```

# Series
DataFrame 에서 하나의 column 만 선택하면 series

```python
# 중복 없는 값들을 나열
print(df_iris.A.unique())

# 중복 없는 값들의 갯수
df_iris.A.nunique()
```

```python
#  각각의 값들이 몇개가 있는지 알 수 있다.
df_iris.A.value_counts()
```

```python
# nlargest 다중 나열
# 해당 column의 상위 n개의 데이터를 가져옴
df_iris.A.nlargest(10)
df_iris.nlargest(10, ["A", "B"])
```

```python
import pandas as pd

# 예제 데이터프레임 생성
data = {
    'CustomerID': [101, 102, 103, 104],
    'OrderDate': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04'],
    'Product': ['Apple', 'Banana', 'Cherry', 'Date'],
    'Quantity': [5, 3, 4, 2],
    'Price': [100,200,250,50]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.rename(columns = {"CustomerID": "ID"}, inplace=True)
```

```python
df.drop(labels="OrderDate", axis=1, inplace=True )
```

```python
df["TotalAmount"] = df.Quantity * df.Price
```
